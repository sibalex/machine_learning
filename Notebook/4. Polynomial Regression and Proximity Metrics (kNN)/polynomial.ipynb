# Полиномиальная регрессия

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = list(range(5))
y = np.sin(x)

plt.plot(x, y, 'o')
```

```python
np.linspace(0, 10, 50)
```

```python
x = list(np.linspace(0, 5, 7))
y = np.sin(x)

plt.plot(x, y, 'o')
```

```python
coefs = np.polyfit(x, y, 2)
coefs
```

```python
np.polyval(coefs, x)
```

```python
plt.plot(x, y, 'o')
plt.plot(x, np.polyval(np.polyfit(x, y, 6), x))
```
