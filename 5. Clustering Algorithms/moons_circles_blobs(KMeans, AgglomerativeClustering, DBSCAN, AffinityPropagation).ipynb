### Делаем множества

```python
from sklearn.datasets import make_moons, make_circles, make_blobs
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
X_moons, Y_moons = make_moons(n_samples=250, noise=0.1)
plt.scatter(X_moons[:,0], X_moons[:,1], c=Y_moons)
```

```python
X_circles, Y_circles = make_circles(n_samples=250, noise=0.1, factor=0.5)
plt.scatter(X_circles[:,0], X_circles[:,1], c=Y_circles)
```

```python
X_blobs, Y_blobs = make_blobs(n_samples=250, centers=3)
plt.scatter(X_blobs[:,0], X_blobs[:,1], c=Y_blobs)
```

```python
plt.scatter(X_blobs[:,0], X_blobs[:,1])
```

### K-Means

```python
from sklearn.cluster import KMeans
```

```python
kmeans_blob = KMeans(n_clusters=3)
out = kmeans_blob.fit_predict(X_blobs)
```

```python
plt.scatter(X_blobs[:,0], X_blobs[:,1], c=out)
```

```python
######
```

```python
kmeans_blob = KMeans(n_clusters=2)
out = kmeans_blob.fit_predict(X_blobs)
plt.scatter(X_blobs[:,0], X_blobs[:,1], c=out)
```

```python
kmeans_blob = KMeans(n_clusters=23)
out = kmeans_blob.fit_predict(X_blobs)
plt.scatter(X_blobs[:,0], X_blobs[:,1], c=out)
```

```python
######
```

Найдем оптимальное колличество кластеров

```python
intertias = []

for k in range(2, 10):
    mns = KMeans(n_clusters=k)
    mns.fit(X_blobs)
    intertias.append(mns.inertia_)
    
plt.plot(list(range(2, 10)), intertias)
```

```python
intertias = []

for k in range(2, 10):
    mns = KMeans(n_clusters=k)
    mns.fit(X_circles)
    intertias.append(mns.inertia_)
    
plt.plot(list(range(2, 10)), intertias)
```

### Агломеративная кластеризация

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
aglo_blobs = AgglomerativeClustering(n_clusters=3)
out = aglo_blobs.fit_predict(X_blobs)
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=out)
```

```python
aglo_circles = AgglomerativeClustering(n_clusters=2)
out = aglo_circles.fit_predict(X_circles)
plt.scatter(X_circles[:, 0], X_circles[:, 1], c=out)
```

```python
aglo_moons = AgglomerativeClustering(n_clusters=2)
out = aglo_moons.fit_predict(X_moons)
plt.scatter(X_moons[:, 0], X_moons[:, 1], c=out)
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
# dbscan_blobs = DBSCAN()

# из минусов надо играть с min_samples и eps
# dbscan_blobs = DBSCAN(eps=1., min_samples=3)

dbscan_blobs = DBSCAN(eps=1.1, min_samples=3)
out = dbscan_blobs.fit_predict(X_blobs)
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=out)
```

```python
dbscan_moons = DBSCAN(eps=0.17, min_samples=3)
out = dbscan_moons.fit_predict(X_moons)
plt.scatter(X_moons[:, 0], X_moons[:, 1], c=out)
```

```python
dbscan_circles = DBSCAN(eps=0.11, min_samples=5)
out = dbscan_circles.fit_predict(X_circles)
plt.scatter(X_circles[:, 0], X_circles[:, 1], c=out)
```

## Affinity Propagation

```python
from sklearn.cluster import AffinityPropagation
```

```python
# можно растить damping

affprop_blobs = AffinityPropagation(damping=0.970)
out = affprop_blobs.fit_predict(X_blobs)
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=out)
```

```python
affprop_circles = AffinityPropagation(damping=0.984)
out = affprop_circles.fit_predict(X_circles)
plt.scatter(X_circles[:, 0], X_circles[:, 1], c=out)
```

```python
affprop_moon = AffinityPropagation(damping=0.984)
out = affprop_moon.fit_predict(X_moons)
plt.scatter(X_moons[:, 0], X_moons[:, 1], c=out)
```

## Метрики

```python
from sklearn.metrics import silhouette_score, adjusted_rand_score
```

```python
kmeans_blob = KMeans(n_clusters=3)
out = kmeans_blob.fit_predict(X_blobs)
plt.scatter(X_blobs[:,0], X_blobs[:,1], c=out)
```

```python
print('ARI: {}'.format(adjusted_rand_score(Y_blobs, out)))
print('SIL: {}'.format(silhouette_score(X_blobs, out)))
```
